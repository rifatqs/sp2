```python
from os.path import abspath
warehouse_location = abspath('spark-warehouse')
scSpark = SparkSession \
    .builder \
    .appName("Spark Hive") \
    .config("spark.sql.warehouse.dir", warehouse_location) \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
scSpark.sql("CREATE DATABASE IF NOT EXISTS sp2")
```

```python
scSpark.sql("SHOW DATABASES").show()
```

```python
scSpark.sql("USE sp2")
```

```python
scSpark.sql("CREATE TABLE IF NOT EXISTS covid \
(location STRING, date DATE, variant STRING, num_sequences INT, perc_sequences FLOAT, num_sequences_total INT) \
tblproperties ('skip.header.line.count'='1') \
row format delimited fields terminated by ','")
```

```python
scSpark.sql("LOAD DATA INPATH \
'/covid-variants.csv' \
into table covid")
```

```python
d = scSpark.sql("select * from covid")
```

```python
d.show()
```

```python
d.count()
```

```python
df = scSpark.sql("SELECT * from covid WHERE date IS NOT NULL")
```

```python
df.show()
```

```python
df.count()
```

```python
df_1 = df.select('location', 'num_sequences') \
.groupBy('location') \
.sum('num_sequences') \
.withColumnRenamed('sum(num_sequences)', 'jumlah_kasus')
```

```python
df_1.show()
```

```python
# Urutan negara dengan total kasus terbanyak
df_1.sort(['jumlah_kasus'], ascending=False).show()
```

```python
df_2 = df.select('variant', 'location', 'num_sequences') \
.groupBy('variant', 'location') \
.sum('num_sequences') \
.withColumnRenamed('sum(num_sequences)', 'jumlah_kasus')
```

```python
df_2.show()
```

```python
# Urutkan negara paling banyak setiap varian
df_2.sort(df_2['variant'].asc(),df_2['jumlah_kasus'].desc()).show(1000)
```

```python
df_3 = df.select('date', 'num_sequences') \
.groupBy('date') \
.sum('num_sequences') \
.withColumnRenamed('sum(num_sequences)', 'jumlah_kasus')
```

```python
df_3.show()
```

```python
# Tanggal puncak kasus terbanyak.
df_3.sort(['jumlah_kasus'], ascending=False).show()
```

```python
scSpark.sql("select location, variant, sum(num_sequences) as total_kasus from covid group by location, variant sort by variant asc, total_kasus desc ").show()
```

```python
scSpark.sql("select location, variant, sum(num_sequences) as total_kasus \
from covid \
group by location, variant \
sort by variant asc, total_kasus desc ").show()
```
